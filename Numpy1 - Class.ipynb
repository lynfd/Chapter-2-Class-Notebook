# Array-Oriented Programming with NumPy

### Python lists are essentially an array of pointers, 
* Each pointing to a location that contains the information related to the element. 
* This adds a lot of overhead in terms of memory and computation. 
* And most of this information is rendered redundant when all the objects stored in the list are of the same type!

### NumPy arrays that contain only homogeneous elements, i.e. elements having the same data type.
* NumPy is a Python library used for working with arrays
* NumPy stands for Numerical Python
* The array object in NumPy is called ndarray
* Many popular data science libraries such as Pandas, SciPy (Scientific Python) and Keras (for deep learning) are built on or depend on NumPy

### 2. Inspection of Numpy array 


```python
!pip install prettytable
```

```python
from prettytable import PrettyTable
def inspect_your_array(array):
    PT=PrettyTable()
    PT.field_names=["Shape","len","Dimension","Size","Datatype","Name of Datatype"]
    PT.add_row([array.shape,len(array),array.ndim,array.size,array.dtype,array.dtype.name])
    print(PT)
```

```python
arr1=np.arange(10)
arr1
```

```python
inspect_your_array(arr1)
```

### Using np.linspace()
* Produce evenly spaced floating-point ranges with NumPy’s **`linspace`** function
* Ending value **is included** in the `array`

```python
data1=np.linspace(1,10,50)
print(data1)
```

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt
plt.plot(data1,np.sin(data1))
plt.title("Sin Curve")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show
```

### Create a cosine signal of start value 1, stop value 50 with 500 linsspace. Do proper label

```python
data2=np.linspace(1,50,500)
import matplotlib.pyplot as plt
plt.plot(data2,np.cos(data2))
plt.title("Cos Curve")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show
```

## Reshaping and Transposing
We often want to reshape or transpose arrays when we are implementing mathematical structures and formulae.  NumPy provides this functionality with some subtleties.

### reshape 
reshape() provides a view into the data - it doesn't change the original array.

```python
data2=np.array([11,46,332,78,54,93,25,27,20,65])
```

```python
data2.size
```

```python
data2.reshape(5,2)
```

### resize
resize() will change the underlaying  array.

```python
data2.resize(5,2)
data2
```

### horizontal and vertical stacking
- np.vstack()
- np.hstack()

```python
data4=np.array([56,78,43,98,90,32,45,12,44,55,88,10])
data3=np.arange(12)
print(data4)
print(data3)
```

```python
n5=np.vstack((data4,data3))
```

```python
print(n5)
```

```python
n4=np.hstack((data4,data3))
print(n4)
```

### Splitting an array
- np.vsplit()
- np.hsplit()
- np.split()

```python
n6,n7=np.hsplit(n4,2)
print(n6)
```

```python
print(n7)
```

```python
n8,n9=np.vsplit(n5,2)
print(n8)
print(n9)
```
